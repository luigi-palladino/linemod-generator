```python
from make_bg import make_bg
import os
import tqdm
import cv2
import random
import skimage.io
import matplotlib.pyplot as plt
import numpy as np
from skimage import img_as_ubyte
```

```python
import pandas as pd

df=pd.read_csv("dataset.csv",index_col=0)
df = df.reset_index()
df
```

```python
images = df["path"]
bgs = os.listdir("D:\\scimmie\\frames4k\\")
dst = ".\\res_jpg\\"

index_img = 0
total_index=index_img
for img in tqdm.tqdm(images):
    index_bg = 0
    try:
        random.shuffle(bgs)
        bg = bgs[0]
        cv2.imwrite(dst+img.split("\\")[-1].replace(".png",".jpg"),make_bg(img,"D:\\scimmie\\frames4k\\"+bg))
    except:
        bg = bgs
        cv2.imwrite(dst+img.split("\\")[-1].replace(".png",".jpg"),cv2.imread(img))
    index_bg+=1
    total_index+=1
    index_img+=1
```

```python
import cv2
x1 = []
y1 = []
x2 = []
y2 = []
for image_path in df["path"]:
    img = skimage.io.imread(image_path.replace('jpg','png'))
    mask = img[:,:,3]
    skimage.io.imsave(dst.replace("jpg","mask")+image_path.split("\\")[-1].replace(".png",".jpg"),mask)
    x, y, w, h = cv2.boundingRect(mask)
    x1.append(x)
    y1.append(y)
    x2.append(w)
    y2.append(h)

    print(x,y,w,h)

df["x1"]= x1
df["y1"]= y1
df["x2"]= x2
df["y2"]= y2

df.to_csv("dataset_with_bbox.csv")    

df
```

```python
import ruamel
from ruamel.yaml import YAML
yaml2 = YAML()
class NonAliasingRTRepresenter(ruamel.yaml.representer.RoundTripRepresenter):
    def ignore_aliases(self, data):
        return True
yaml2.Representer = NonAliasingRTRepresenter
yaml2.default_flow_style = True

with open('new_gt.yml', 'w') as outfile:
    yaml2.dump(dicts, outfile)

# with open('gt.yml') as f:
#     data = yaml.load(f)
#     print(data)
```
